```python
import easyocr

# 创建easyocr读取器，指定语言为英文
reader = easyocr.Reader(['en'])

# 读取图片中的文本
results = reader.readtext('C:/Users/zouxu/OneDrive/Desktop/p poker/dataset/face/4.png')

# 用于存储最终结果的列表
final_results = []

# 遍历识别结果
for result in results:
    # 获取文本的四个坐标点
    points = result[0]
    # 计算中心点坐标
    center_x = sum([point[0] for point in points]) / 4
    center_y = sum([point[1] for point in points]) / 4
    # 获取文本内容
    text = result[1]
    # 添加到最终结果列表中
    final_results.append(((center_x, center_y), text))

# 打印最终结果
for res in final_results:
    print(res)
```

## 从视频抽取图片模块

```python
import cv2
import os
from datetime import timedelta

video_path = r'C:\Users\zouxu\OneDrive\Desktop\p poker\dataset\video\10_30_23.mp4'
output_folder = r'C:\Users\zouxu\OneDrive\Desktop\p poker\dataset\test8'

# 确保输出文件夹存在
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 准备读取视频
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)  # 获取帧率
interval = 10 * fps  # 每5秒提取一帧

# 加载面部检测器
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 处理视频
frameId = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 每5秒提取一帧
    if frameId % interval == 0:
        # 在图片中检测面部
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)

        # 检查面部数量是否为1或2
        if len(faces) == 1 or len(faces) == 2:
            timestamp = str(timedelta(seconds=frameId/fps)).replace(":", "-")
            image_path = os.path.join(output_folder, f"{timestamp}.jpg")
            cv2.imwrite(image_path, frame)  # 将帧保存为图片

    frameId += 1

cap.release()

print("处理完成")

```

## 整合模块： 1.初选图片（只保留面部为1、2，文字内容大于50）2.保留文字信息

```python
import os
import csv
import face_recognition
import easyocr

def process_images_and_save_text(directory_path, output_csv_path):
    # 创建一个easyocr reader实例
    reader = easyocr.Reader(['en'])  # 假设语言是英文

    # 准备写入CSV文件
    with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        filewriter = csv.writer(csvfile)
        filewriter.writerow(['id', 'text'])  # 写入标题行
        
        # 遍历指定目录下的所有文件
        for filename in os.listdir(directory_path):
            if filename.endswith(".jpg") or filename.endswith(".png"): # 仅处理jpg和png图片
                file_path = os.path.join(directory_path, filename)
                
                # 加载图片并识别面部
                image = face_recognition.load_image_file(file_path)
                face_locations = face_recognition.face_locations(image)
                num_faces = len(face_locations)
                
                # 使用easyocr读取图片中的文字
                text_results = reader.readtext(file_path)
                text = ' '.join([result[1] for result in text_results])

                # 检查面部数量和文字数量，进行相应处理
                if num_faces not in [1, 2] or len(text) < 50:
                    os.remove(file_path)
                    print(f"{filename} was deleted because it does not meet the criteria.")
                else:
                    # 将符合条件的文件名（id）和识别出的文本（text）写入CSV
                    filewriter.writerow([filename, text])
                    print(f"Processed {filename}: found {num_faces} face(s) and text successfully saved.")

if __name__ == "__main__":
    directory_path =  "C:/Users/zouxu/OneDrive/Desktop/p poker/dataset/test4"
    output_csv_path = "C:/Users/zouxu/OneDrive/Desktop/p poker/dataset/test4/text_data.csv"
    process_images_and_save_text(directory_path, output_csv_path)

    print("Processing complete. Text data saved to CSV.")
```

```python
import os
import csv
import face_recognition
import easyocr

def process_images_and_save_text(directory_path, output_csv_path):
    # 创建一个easyocr reader实例
    reader = easyocr.Reader(['en'])  # 假设语言是英文

    # 准备写入CSV文件
    with open(output_csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        filewriter = csv.writer(csvfile)
        filewriter.writerow(['id', 'text'])  # 写入标题行
        
        # 遍历指定目录下的所有文件
        for filename in os.listdir(directory_path):
            if filename.endswith(".jpg") or filename.endswith(".png"): # 仅处理jpg和png图片
                file_path = os.path.join(directory_path, filename)
                
                # 加载图片并识别面部
                image = face_recognition.load_image_file(file_path)
                face_locations = face_recognition.face_locations(image)
                num_faces = len(face_locations)

                # 使用easyocr读取图片中的文字，并计算每个文本段的中心坐标
                text_results = reader.readtext(file_path)
                
                # 生成文本和位置信息的字符串
                text_info = '; '.join([f"(({int(sum([point[0] for point in result[0]]) / 4)}, {int(sum([point[1] for point in result[0]]) / 4)}), {result[1]})" for result in text_results])

                # 检查面部数量，进行相应处理
                if num_faces not in [1, 2] or len(text_info) < 50:
                    os.remove(file_path)
                    print(f"{filename} was deleted because it does not meet the criteria.")
                else:
                    # 将每个图片的所有文本信息作为一行数据写入CSV
                    filewriter.writerow([filename, text_info])
                    print(f"Processed {filename}: found {num_faces} face(s) and text successfully saved.")

if __name__ == "__main__":
    directory_path = "C:/Users/zouxu/OneDrive/Desktop/p poker/dataset/test8"
    output_csv_path = "C:/Users/zouxu/OneDrive/Desktop/p poker/dataset/test8/text_data1.csv"
    process_images_and_save_text(directory_path, output_csv_path)

    print("Processing complete. Text data saved to CSV.")

```
