## steps of HOG

```python
import dlib
from skimage import io
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage.color import rgb2gray
```

```python
file_name = r"C:\Users\zouxu\OneDrive\Desktop\p poker\sum sprint\6_1\1.jpg"

image = io.imread(file_name)
gray_image = rgb2gray(image)

# HOG
face_detector = dlib.get_frontal_face_detector()
detected_faces = face_detector(image, 1)

# 
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.imshow(image)
for face in detected_faces:
    rect = plt.Rectangle((face.left(), face.top()), face.width(), face.height(),
                         edgecolor='r', facecolor='none')
    ax.add_patch(rect)
plt.title('Detected Faces')
plt.show()
```

```python
# HOGHOG
fd, hog_image = hog(gray_image, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True)

# HOG
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].imshow(gray_image, cmap=plt.cm.gray)
ax[0].set_title('Input Image')
ax[1].imshow(hog_image, cmap=plt.cm.gray)
ax[1].set_title('HOG Features')
plt.show()
```
